```python
import matplotlib.pyplot as plt
import numpy as np

years = ['2019', '2020', '2021', '2022', '2023 YTD']
funding = [1.2, 0.5, 3.5, 2.5, 14.1]  
deals = [59, 89, 137, 141, 86]

# Calculate linear regression
x = np.array([2019, 2020, 2021, 2022, 2023])
y_fund = np.array(funding)
y_deal = np.array(deals)

m_fund, b_fund = np.polyfit(x, y_fund, 1) 
m_deal, b_deal = np.polyfit(x, y_deal, 1)

# Make predictions
x_future = np.arange(2024, 2026)
y_fund_pred = m_fund * x_future + b_fund
y_deal_pred = m_deal * x_future + b_deal

# Plot
fig, ax1 = plt.subplots(figsize=(10,5))

ax1.bar(years, funding, color='skyblue', alpha=0.6)
ax1.plot(np.append(years, x_future.astype(str)), np.append(y_fund, y_fund_pred), 'tab:blue')
ax1.set_ylim(0, max(np.append(y_fund, y_fund_pred)) + 2)
ax1.set_xlabel('Year')
ax1.set_ylabel('Funding ($B)', color='skyblue')

ax2 = ax1.twinx()  
ax2.plot(years, deals, 'tab:blue', marker='o')
ax2.plot(np.append(years, x_future.astype(str)), np.append(y_deal, y_deal_pred), 'tab:blue') 
ax2.set_ylim(0, max(np.append(y_deal, y_deal_pred)) + 20)
ax2.set_ylabel('Deals', color='tab:blue')

plt.title('Yearly Funding and Deals (Projected to 2025)')
plt.tight_layout()
plt.show()
```
